```python
# Import pandas and numpy, DataFrame / Series
import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

#Question 1 asks that we identify the northernmost airport in the United States from the below imported airports.csv file
#To do so we must first import the file from https://github.com/tidyverse/nycflights13/blob/main/data-raw/airports.csv

US_Airports = pd.read_csv('https://raw.githubusercontent.com/tidyverse/nycflights13/main/data-raw/airports.csv')
US_Airports.head()
```

```python
#Sort the values in the column "latitude" of the imported dataset in descending order to idenitfy the top 5 northernmost airports on the list.
US_Airports.sort_values(by=['lat'], inplace=True, ascending=False)
US_Airports.head(5)

#Because Alaska is one of the northernmost states in the United States in the US, it is no surprise that the top 5 northernmost airports are located in Alaska
#Nonethless, there seems to be an error on the list, the Dillant Hopkins Airport's timezone is missing and the coordinates are incorrectly listed.
# After a Google search it was confirmed that:
# 1. The Dilliant Hopkins Airport is located in Swanzey, New Hampshire
# 2. The timezone is America/New_York
# 3. The coordinates are: Lat 42.9065° N, Long: -72.2726° W
#Reference (Location Data): https://en.wikipedia.org/wiki/Dillant%E2%80%93Hopkins_Airport
#Refernce (Time Zone: https://www.prokerala.com/travel/airports/united-states-of-america/dillant-hopkins-airport.html)
```

```python
# Correct the coordinates entries for the Dilliant Hopkins Airport and display the results
US_Airports.loc[417,['lat','lon', 'tzone']] = [42.9065,-72.2726,'America/New_York']
US_Airports.loc[ [417] , : ]
```

```python
# Display the sorted tables with the updated values

US_Airports.sort_values(by=['lat'], inplace=True, ascending=False)
US_Airports.head(5)

# After we sucessfully updated the values for the Dillant Hopkins Airport the coordinates.
#I validated that the top 5 northernmost airports on the list and their coordinates are correctly listed.
```

```python
# Therefore:
# Question 1: What is the northernmost airport in the United States?
# Answer: The northernmost airport in the United States is the Wiley Post Will Rogers Memorial Airport

Northernmost_aiportUS = US_Airports.loc[ [230] , : ]
Northernmost_aiportUS
```

```python
# Question 3 asks that we identify which airport in the New York area that had the windiest weather on  February 12th, 2013.
# To do so we must first import the file from https://raw.githubusercontent.com/tidyverse/nycflights13/main/data-raw/weather.csv

NYArea_AirportWeather = pd.read_csv('https://raw.githubusercontent.com/tidyverse/nycflights13/main/data-raw/weather.csv')
NYArea_AirportWeather.head()
```

```python
# Split the "time_hour" column to create two new columns; date & time

NYArea_AirportWeather[['date', 'time']] = NYArea_AirportWeather['time_hour'].str.split('T', 1, expand=True)
NYArea_AirportWeather.head()
```

```python
# Remove the "time_hour" column

NYArea_AirportWeather.drop(NYArea_AirportWeather.columns[[14]], axis = 1, inplace = True)
NYArea_AirportWeather.head()
```

```python
# Filter the DataFrame by the desired date: February 13, 2013.

NYArea_AirportWeather = NYArea_AirportWeather[(NYArea_AirportWeather['date'] >= '2013-02-13') & (NYArea_AirportWeather['date'] <= '2013-02-13')]
NYArea_AirportWeather
```

```python
# Sort the column "wind-speed" in descending order to indentify the airport with the greatest wind speed

NYArea_AirportWeather.sort_values(by=['wind_speed'], inplace=True, ascending=False)
NYArea_AirportWeather
```

```python
# # Therefore:
# Question 3: On February 12th, 2013, which New York area airport had the windiest weather?
# Answer: Based on the data in the table, the  airport in the New York Area  that experinced the windiest weather is LGA

Windiest_NYAirport = NYArea_AirportWeather.loc[ [18434] , : ]
Windiest_NYAirport
```
