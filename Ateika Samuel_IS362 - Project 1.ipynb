```python
# Import pandas and numpy, DataFrame / Series

import pandas as pd
import numpy as np
from pandas import DataFrame, Series

# Add items for matplotlib
%matplotlib inline
import matplotlib
matplotlib.style.use('ggplot')

#Read the contents of Ateika_Samuel_IS362_Project1.csv into a DataFrame

airlines_datafile = pd.read_csv("D:/IS362_Project1/Ateika_Samuel_IS362_Project1.csv")

#Display the contents of Ateika_Samuel_IS362_Project1.csv DataFrame
airlines_datafile
```

```python
# Sum the two columns Alaska: On time flights & Alaska: Delayed flights to get the Total flights for Alaska
airlines_datafile["Alaska's Total Flights"] = airlines_datafile["Alaska: On time flights"] + airlines_datafile["Alaska: Delayed flights"]

# Calculate the percentage of delayed flights for Alaska
airlines_datafile["Alaska's: Delayed flights Percentage"] = (airlines_datafile["Alaska: Delayed flights"] / airlines_datafile["Alaska's Total Flights"]) * 100

#Display the new contents of Ateika_Samuel_IS362_Project1.csv DataFrame
airlines_datafile
```

```python
# Display the graph from the DataFrame for Alaska's delayed flights percentage
Pie_chart_labels=['Los Angeles','Phoenix','San Diego','San Francisco','Seattle']
airlines_datafile
airlines_datafile.plot.pie(labels=Pie_chart_labels, y="Alaska's: Delayed flights Percentage")
```

```python
# Calculate the percentage of On time flights for Alaska
airlines_datafile["Alaska's: On time flights Percentage"] = (airlines_datafile["Alaska: On time flights"] / airlines_datafile["Alaska's Total Flights"]) * 100

#Display the new contents of Ateika_Samuel_IS362_Project1.csv DataFrame
airlines_datafile
```

```python
# Display the graph from the DataFrame for Alaska's On time flights percentage
Pie_chart_labels=['Los Angeles','Phoenix','San Diego','San Francisco','Seattle']
airlines_datafile
airlines_datafile.plot.pie(labels=Pie_chart_labels, y="Alaska's: On time flights Percentage")
```

```python
# Sum the two columns AM WEST: On time flights & AM WEST: Delayed flights to get the Total flights for AM WEST
airlines_datafile["AM WEST's Total Flights"] = airlines_datafile["AM WEST: On time flights"] + airlines_datafile["AM WEST: Delayed flights"]

# Calculate the percentage of delayed flights for AM WEST
airlines_datafile["AM WEST's: Delayed flights Percentage"] = (airlines_datafile["AM WEST: Delayed flights"] / airlines_datafile["AM WEST's Total Flights"]) * 100

#Display the new contents of Ateika_Samuel_IS362_Project1.csv DataFrame
airlines_datafile
```

```python
# Display the graph from the DataFrame for AM WEST's delayed flights percentage
Pie_chart_labels=['Los Angeles','Phoenix','San Diego','San Francisco','Seattle']
airlines_datafile
airlines_datafile.plot.pie(labels=Pie_chart_labels, y="AM WEST's: Delayed flights Percentage")
```

```python
# Calculate the percentage of On time flights for AM WEST
airlines_datafile["AM WEST's: On time flights Percentage"] = (airlines_datafile["AM WEST: On time flights"] / airlines_datafile["AM WEST's Total Flights"]) * 100

#Display the new contents of Ateika_Samuel_IS362_Project1.csv DataFrame
airlines_datafile
```

```python
# Display the graph from the DataFrame for AM WEST's on time flights percentage
Pie_chart_labels=['Los Angeles','Phoenix','San Diego','San Francisco','Seattle']
airlines_datafile
airlines_datafile.plot.pie(labels=Pie_chart_labels, y="AM WEST's: On time flights Percentage")
```

```python
#Data Comparisons: compare the data to confirm which Airline has the most delayed flights
for index, row in airlines_datafile.iterrows():
    if [row["Alaska's: Delayed flights Percentage"]] > [row["AM WEST's: Delayed flights Percentage"]]:
        airlines_datafile.loc[index, 'Most delayed flights by Airlines'] = 'Alaska'
    else:
        airlines_datafile.loc[index,'Most delayed flights by Airlines'] = 'AMWest'

airlines_datafile['Most delayed flights by Airlines']
```
